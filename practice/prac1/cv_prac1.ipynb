# Лабораторная работа №1. Создание типа данных «класс»
1. Создать класс с полями, указанными в индивидуальном задании.
2. Реализовать в классе методы:
    * конструктор по умолчанию;
    * деструктор для освобождения памяти (с сообщением об уничтожении
    объекта);
    * функции обработки данных, указанные в индивидуальном задании;
    * функцию формирования строки информации об объекте.
3. Создать проект для демонстрации работы: сформировать объекты со значениями-константами и с введенными с клавиатуры значениями полей объекта. В основной ветке программы создайте три объекта класса. Вывести результаты работы на экран.


| № варианта | Класс-родитель и его поля | Функция-метод 1 обработки данных | Функция-метод 2 обработки данных |
|-|-|-|-|
|11 | Работник: фамилия, должность, оклад | Увеличить оклад на 15% (каждому работнику) | Работникам, у которых фамилия начинается с сочетания букв «Иван», присвоить должность «инженер»

```python
class Worker:
    def __init__(self, surname="", position="", salary=0):
        """
        Конструктор по умолчанию.
        Инициализирует объект работника.
        """
        self.surname = surname
        self.position = position
        self.salary = float(salary)
        print(f"Создан работник: {self.surname}")

    def __del__(self):
        """
        Деструктор — вызывается при удалении объекта.
        Оповещает об уничтожении объекта.
        """
        print(f"Объект Worker ({self.surname}) уничтожен. Память освобождена.")

    def increase_salary(self):
        """
        Увеличивает оклад на 15%.
        """
        self.salary *= 1.15

    def assign_engineer_position(self):
        """
        Если фамилия начинается с 'Иван', присваивает должность 'инженер'.
        """
        if self.surname.startswith("Иван"):
            self.position = "инженер"

    def get_info(self):
        """
        Возвращает строку с информацией о работнике.
        """
        return f"Фамилия: {self.surname}, Должность: {self.position}, Оклад: {self.salary:.2f}"
```

---
Объект 1: значения-константы
---

```python
worker1 = Worker("Иванов", "менеджер", 50000)
```

---
Объект 2: значения-константы
---

```python
worker2 = Worker("Петров", "программист", 60000)
```

---
Объект 3: ввод с клавиатуры
---

```python
print("Введите данные для третьего работника:")
surname = input("Фамилия: ")
position = input("Должность: ")
try:
    salary = float(input("Оклад: "))
except ValueError:
    salary = 0.0
    print("Некорректный ввод оклада. Установлено значение 0.0")
worker3 = Worker(surname, position, salary)
```

---
Список всех работников
---

```python
workers = [worker1, worker2, worker3]
```

---
Данные до обработки
---

```python
for i, worker in enumerate(workers, start=1):
    print(f"Работник {i}: {worker.get_info()}")
```

---
Применение обработки данных
---

```python
for worker in workers:
    worker.increase_salary()           # Увеличить оклад на 15%
    worker.assign_engineer_position()  # Назначить должность инженер, если фамилия начинается с "Иван"
```

---
Данные после обработки
---

```python
for i, worker in enumerate(workers, start=1):
    print(f"Работник {i}: {worker.get_info()}")
```

---
Удаление объектов
---

```python
del worker1
del worker2
del worker3
```
